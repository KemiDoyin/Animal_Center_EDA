```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, time
%matplotlib inline


plt.style.use('ggplot')
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (11,8)
```

```python
#import and read csv

df = pd.read_csv('C:/Users/HP/Downloads/Austin_Animal_Center_Intakes.csv/Austin_Animal_Center_Intakes.csv')
df
```

```python
#Remove unwanted character
#Fill empty cells
#Delete Unwanted Columns

df.Name= df.Name.str.strip('*')
df.fillna('Unknown',inplace=True)

del_col = ['Animal ID', 'MonthYear']

df.drop(del_col, axis=1, inplace=True)
```

```python
#Convert to datetime

df['DateTime'] = pd.to_datetime(df['DateTime'], format= '%m/%d/%Y %I:%M:%S %p')
```

```python
#Extract Sex into a new column

df['New_Sex'] = df['Sex upon Intake'].str.extract(r'(Male|Female|Unknown)')
```

```python
#Plot a scatter plot to show Intake type by Animal Type

sns.scatterplot(
    data=df, x="Intake Type", y="Animal Type", hue="New_Sex",
    sizes=(20, 200), legend="full"
)
plt.show()
```

```python
#Check each values for unwanted characters

df['Age upon Intake'].unique()  
```

```python
#Remove unwanted character from the string

df['Age upon Intake'] = df['Age upon Intake'].str.strip('-')
```

```python
#Convert the age upon intake to weeks and extract into a new column for visualization

def convert_age_to_weeks(age_str):
    if 'week' in age_str:
        return int(age_str.split()[0])
    elif 'month' in age_str:
        months = int(age_str.split()[0])
        return int(months * 4.33)  # Approximation: 1 month â‰ˆ 4.33 weeks
    elif 'year' in age_str:
        years = int(age_str.split()[0])
        return int(years * 52) # Approximation: 1 year â‰ˆ 52 weeks
    elif 'day' in age_str:
        days = int(age_str.split()[0])
        return int(days / 7) #Convert days to weeks
    else:
        return None  # Handle unexpected formats

df['Age_in_weeks'] = df['Age upon Intake'].apply(convert_age_to_weeks)
```

```python
#round the values of age in weeks to 1 and change the type to integer

df['Age_in_weeks'] =df['Age_in_weeks'].round().astype(int)

df.dtypes
```

```python
#Number of each Intake type by their gender

ct = pd.crosstab(df['Intake Type'], df['New_Sex'])

# Create a stacked bar chart

ct.T.plot(kind='bar', stacked=True)


# Customize the chart (optional)
plt.title("Sex by Intake Type")
plt.xlabel("Intake Type", color='red')
plt.ylabel("Count", color='red')
plt.xticks(rotation=0)

plt.show()
```

```python
#Age of animals in weeks by their gender. 200 above means >= 3 years of age

custom_colors = {'Male': 'Blue', 'Female': 'Green', 'Unknown': 'Red'}

sns.violinplot(data=df, x='New_Sex', y='Age_in_weeks', palette=custom_colors)



plt.title('Age in weeks by their gender')
plt.xlabel("Sex", color='orange')
plt.ylabel("Age in weeks", color='orange')

plt.show()
```

```python
#Total number of animals in shelter

Animals = df['Animal Type'].value_counts()

Animals.plot(kind='bar')

plt.xlabel('Type of Animals')
plt.ylabel('Count')
plt.title('Total Number of Animals in Shelter')
```

```python
#Total number of animals taken in by condition type

condition_counts = {} 
for condition in df['Intake Condition']:
    if condition in condition_counts:
        condition_counts[condition] += 1
    else:
        condition_counts[condition] = 1


condition_labels = list(condition_counts.keys())
condition_values = list(condition_counts.values())



sns.barplot(x=condition_values, y=condition_labels)

plt.xlabel('Total number', color='red')
plt.ylabel('Condition Type', color='red')
plt.title('Animals Condition')

plt.show()
```

```python
#Get the total number of each animal type brought into shelter yearly

#create a new column to extract 'year' from 'datetime'

df['Year'] = df['DateTime'].dt.year

#group by year and animal type

df.groupby(['Year','Animal Type']).size()

```

```python

#Create a time series to show the total intake of each animal yearly

animal_counts = df.groupby(['Year', 'Animal Type']).size().reset_index(name='Count')

#Plot a time series

sns.lineplot(data=animal_counts, x='Year', y='Count', hue='Animal Type')

plt.xlabel('Year', color='green')
plt.ylabel('Animal Count', color='green')
plt.title('Animals Taken into Shelter')




plt.show()
```
